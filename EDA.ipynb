## Introduction

Sleep is a fundamental biological process essential for maintaining physical health, cognitive function, and emotional well-being. According to the National Sleep Foundation and numerous scientific studies, adults typically require between 7 and 9 hours of sleep per night to support optimal health, though individual needs may vary. Chronic sleep deprivation has been linked to a range of adverse health outcomes, including impaired immune function, increased risk of cardiovascular disease, metabolic disorders such as obesity and diabetes, and mental health issues like depression and anxiety (Hirshkowitz et al., 2015; Medic et al., 2017).

The quality of sleep is influenced by a complex interplay of factors, including age, gender, lifestyle habits, physical activity, dietary patterns, and the presence of sleep disorders. For example, research shows that sleep quality tends to decline with age due to changes in circadian rhythms and sleep architecture (Ohayon et al., 2004). Gender differences have also been observed, with women often reporting more sleep disturbances, possibly due to hormonal fluctuations and life events such as pregnancy and menopause (Mallampalli & Carter, 2014).

Physical activity is widely recognized as a protective factor for sleep health. Regular moderate-to-vigorous exercise has been shown to improve sleep quality, increase sleep duration, and reduce the time it takes to fall asleep (Kredlow et al., 2015). Conversely, sedentary behavior and low physical activity levels are associated with poorer sleep outcomes.

Dietary habits also play a significant role in sleep regulation. Diets rich in fruits, vegetables, whole grains, and lean proteins are associated with better sleep quality, while high intake of sugar, saturated fats, and processed foods can disrupt sleep patterns (St-Onge et al., 2016). Additionally, the timing of meals and caffeine or alcohol consumption can impact sleep onset and maintenance.

Sleep disorders, such as insomnia, sleep apnea, and restless legs syndrome, are prevalent conditions that significantly impair sleep quality and overall health. The use of medications to manage sleep disorders is common, but these interventions may not always lead to restorative sleep and can sometimes result in dependency or side effects (Morin et al., 2015).

This project presents an exploratory data analysis (EDA) of a synthetic dataset designed to reflect the multifaceted nature of sleep health. The dataset includes variables such as age, gender, sleep quality, bedtime and wake-up times, daily steps, calories burned, physical activity level, dietary habits, presence of sleep disorders, and medication usage. The primary objective is to investigate how these factors interact and contribute to sleep quality, drawing on scientific literature to contextualize the findings.

By examining these relationships, the analysis aims to provide insights that align with current scientific understanding and highlight the importance of holistic lifestyle approaches for promoting healthy sleep. Although the dataset is fictional and intended for educational purposes, the analytical methods and interpretations are grounded in evidence-based research, demonstrating the value of data-driven approaches in health science.

### The effect of age on sleep quality

```python
print(df[['Age', 'Sleep Quality']].isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Sleep Quality', data=df, alpha=0.5)
plt.title('Effect of Age on Sleep Quality')
plt.xlabel('Age')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
age_sleep_quality_corr = df[['Age', 'Sleep Quality']].corr().iloc[0, 1]
print(f'Correlation coefficient between Age and Sleep Quality: {age_sleep_quality_corr:.2f}')
```

```python
df_clean = df.dropna(subset=['Age', 'Sleep Quality'])

X = df_clean[['Age']]
y = df_clean['Sleep Quality']

model = LinearRegression()
model.fit(X, y)

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Sleep Quality', data=df_clean, alpha=0.5, label='Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.title('Linear Regression between Age and Sleep Quality')
plt.xlabel('Age')
plt.ylabel('Sleep Quality')
plt.legend()
plt.show()

print(f'Regression coefficient: {model.coef_[0]:.2f}')
print(f'Intercept: {model.intercept_:.2f}')
```

### Compare sleep quality and physical activity level between males and females.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Gender', y='Sleep Quality', data=df)
plt.title('Comparison of Sleep Quality Between Males and Females')
plt.xlabel('Gender')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
activity_levels = {'low': 1, 'medium': 2, 'high': 3}
df['Physical Activity Level Numeric'] = df['Physical Activity Level'].map(activity_levels)

plt.figure(figsize=(12, 6))
sns.barplot(x='Gender', y='Physical Activity Level Numeric', data=df)
plt.title('Comparison of Physical Activity Level Between Males and Females')
plt.xlabel('Gender')
plt.ylabel('Physical Activity Level')
plt.show()
```

```python
male_sleep_quality = df[df['Gender'] == 'm']['Sleep Quality']
female_sleep_quality = df[df['Gender'] == 'f']['Sleep Quality']

t_stat_sleep_quality, p_val_sleep_quality = stats.ttest_ind(male_sleep_quality, female_sleep_quality, nan_policy='omit')
print(f'T-statistic for Sleep Quality: {t_stat_sleep_quality:.2f}')
print(f'P-value for Sleep Quality: {p_val_sleep_quality:.4f}')

male_activity_level = df[df['Gender'] == 'm']['Physical Activity Level Numeric']
female_activity_level = df[df['Gender'] == 'f']['Physical Activity Level Numeric']

t_stat_activity_level, p_val_activity_level = stats.ttest_ind(male_activity_level, female_activity_level, nan_policy='omit')
print(f'T-statistic for Physical Activity Level: {t_stat_activity_level:.2f}')
print(f'P-value for Physical Activity Level: {p_val_activity_level:.4f}')
```

### Analyze how the timing of sleep and wakefulness affects sleep quality.

```python
def time_to_hours(time_str):
    hours, minutes = map(int, time_str.split(':'))
    return hours + minutes / 60

df['Bedtime Hours'] = df['Bedtime'].apply(time_to_hours)
df['Wake-up Hours'] = df['Wake-up Time'].apply(time_to_hours)

# Calculate sleep duration in hours
df['Sleep Duration Hours'] = (df['Wake-up Hours'] - df['Bedtime Hours']) % 24
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Bedtime Hours', y='Sleep Quality', data=df, alpha=0.5)
plt.title('Relationship Between Bedtime and Sleep Quality')
plt.xlabel('Bedtime (Hours)')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Wake-up Hours', y='Sleep Quality', data=df, alpha=0.5)
plt.title('Relationship Between Wake-up Time and Sleep Quality')
plt.xlabel('Wake-up Time (Hours)')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sleep Duration Hours', y='Sleep Quality', data=df, alpha=0.5)
plt.title('Relationship Between Sleep Duration and Sleep Quality')
plt.xlabel('Sleep Duration (Hours)')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
df_bedtime = df[['Bedtime Hours', 'Sleep Quality']].dropna()
X_bedtime = df_bedtime[['Bedtime Hours']]
y_bedtime = df_bedtime['Sleep Quality']

model_bedtime = LinearRegression()
model_bedtime.fit(X_bedtime, y_bedtime)
y_pred_bedtime = model_bedtime.predict(X_bedtime)

print(f'R-squared for Bedtime Model: {r2_score(y_bedtime, y_pred_bedtime):.2f}')
```

```python
df_wakeup = df[['Wake-up Hours', 'Sleep Quality']].dropna()
X_wakeup = df_wakeup[['Wake-up Hours']]
y_wakeup = df_wakeup['Sleep Quality']

model_wakeup = LinearRegression()
model_wakeup.fit(X_wakeup, y_wakeup)
y_pred_wakeup = model_wakeup.predict(X_wakeup)

print(f'R-squared for Wake-up Time Model: {r2_score(y_wakeup, y_pred_wakeup):.2f}')
```

```python
df_sleep_duration = df[['Sleep Duration Hours', 'Sleep Quality']].dropna()
X_sleep_duration = df_sleep_duration[['Sleep Duration Hours']]
y_sleep_duration = df_sleep_duration['Sleep Quality']

model_sleep_duration = LinearRegression()
model_sleep_duration.fit(X_sleep_duration, y_sleep_duration)
y_pred_sleep_duration = model_sleep_duration.predict(X_sleep_duration)

print(f'R-squared for Sleep Duration Model: {r2_score(y_sleep_duration, y_pred_sleep_duration):.2f}')
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x='Bedtime Hours', y='Sleep Quality', data=df, alpha=0.5, label='Data')
plt.plot(df_bedtime[['Bedtime Hours']], y_pred_bedtime, color='red', label='Regression Line')
plt.title('Bedtime vs. Sleep Quality')
plt.xlabel('Bedtime (Hours)')
plt.ylabel('Sleep Quality')
plt.legend()


plt.subplot(1, 3, 2)
sns.scatterplot(x='Wake-up Hours', y='Sleep Quality', data=df, alpha=0.5, label='Data')
plt.plot(df_wakeup[['Wake-up Hours']], y_pred_wakeup, color='red', label='Regression Line')
plt.title('Wake-up Time vs. Sleep Quality')
plt.xlabel('Wake-up Time (Hours)')
plt.ylabel('Sleep Quality')
plt.legend()

plt.subplot(1, 3, 3)
sns.scatterplot(x='Sleep Duration Hours', y='Sleep Quality', data=df, alpha=0.5, label='Data')
plt.plot(df_sleep_duration[['Sleep Duration Hours']], y_pred_sleep_duration, color='red', label='Regression Line')
plt.title('Sleep Duration vs. Sleep Quality')
plt.xlabel('Sleep Duration (Hours)')
plt.ylabel('Sleep Quality')
plt.legend()

plt.tight_layout()
plt.show()
```

### Analyze the relationship between sleep duration and sleep quality.

```python
df['Bedtime'] = pd.to_datetime(df['Bedtime'], format='%H:%M').dt.hour + pd.to_datetime(df['Bedtime'], format='%H:%M').dt.minute / 60
df['Wake-up Time'] = pd.to_datetime(df['Wake-up Time'], format='%H:%M').dt.hour + pd.to_datetime(df['Wake-up Time'], format='%H:%M').dt.minute / 60

df['Sleep Duration'] = df['Wake-up Time'] - df['Bedtime']
df['Sleep Duration'] = df['Sleep Duration'].apply(lambda x: x + 24 if x < 0 else x)
```

```python
df_reg = df[['Sleep Duration', 'Sleep Quality']].dropna()
X_sleep_duration = df_reg[['Sleep Duration']]
y_sleep_quality = df_reg['Sleep Quality']

model_sleep_duration = LinearRegression()
model_sleep_duration.fit(X_sleep_duration, y_sleep_quality)
y_pred_sleep_duration = model_sleep_duration.predict(X_sleep_duration)

print(f'R-squared for Sleep Duration Model: {r2_score(y_sleep_quality, y_pred_sleep_duration):.2f}')
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Sleep Duration', y='Sleep Quality', data=df, alpha=0.5, label='Data')
plt.plot(df_reg[['Sleep Duration']], y_pred_sleep_duration, color='red', label='Regression Line')
plt.title('Regression Line for Sleep Duration vs. Sleep Quality')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Sleep Quality')
plt.legend()
plt.show()
```

Individuals who go to bed earlier and maintain regular wake-up times generally report better sleep quality. Sleep duration is positively associated with sleep quality, highlighting the importance of sufficient and consistent rest.

Higher levels of physical activity are linked to better sleep quality. ANOVA results confirmed significant differences in sleep quality across physical activity levels, emphasizing the role of an active lifestyle in promoting restorative sleep.

**5. Calories Burned:**  
There is a moderate positive relationship between daily calories burned and sleep quality. This suggests that greater energy expenditure, often a result of physical activity, may contribute to improved sleep.

**6. Dietary Habits:**  
Participants with healthy dietary habits tend to have higher sleep quality compared to those with medium or unhealthy diets. Statistical analysis confirmed that nutrition is an important factor influencing sleep.

**7. Sleep Disorders and Medication Usage:**  
The presence of sleep disorders is strongly associated with lower sleep quality and reduced physical activity levels. Additionally, individuals using medication for sleep tend to report poorer sleep quality, indicating ongoing sleep challenges despite intervention.

**8. Key Factors Affecting Sleep Quality:**  
Correlation and multiple regression analyses identified sleep duration, presence of sleep disorders, physical activity level, and dietary habits as the most influential factors on sleep quality. Age and medication usage also play a role, but to a lesser extent.
